```python
import pandas as pd
import ast
from dateutil.parser import parse
from datetime import datetime
```

```python
print('*** Video dataset', '\nstart\t', datetime.now())
```

```python
df1 = pd.read_csv('C:/Users/Esteban Guerrero/SocialNetworkProject/Iteration2/Datasets/NetworkFeedRawOutput.csv',
                  index_col = False, encoding = 'ISO-8859-1')
```

```python
vidDF = df1[df1.ExtrType.str.contains('VideoInfo')].reset_index(drop = True)
```

```python
vidDF['Plattform'] = ''
vidDF['Plattform'] = vidDF['ExtrType'].str[:7]
vidDF['ExtractionDateTime'] = vidDF['ExtractionDateTime'].str[:10]
```

```python
vidDF = vidDF.sort_values(by = ['Key2', 'Key1', 'ExtractionDateTime'], ascending = True).reset_index(drop = True)
```

```python
vidDF.head()
```

```python
vidDF['User'] = ''
vidDF['VidDet_publishedAt'] = ''
vidDF['VidDet_title'] = ''
vidDF['VidStats_viewCount'] = ''
vidDF['VidStats_likeCount'] = ''
vidDF['VidStats_dislikeCount'] = ''
vidDF['VidStats_favoriteCount'] = ''
vidDF['VidStats_commentCount'] = ''
```

```python
OutpInd = vidDF.columns.get_loc('Output1')

UserInd = vidDF.columns.get_loc('User')
VidDet_publishedAtInd = vidDF.columns.get_loc('VidDet_publishedAt')
VidDet_titleInd = vidDF.columns.get_loc('VidDet_title')
VidStats_viewCountInd = vidDF.columns.get_loc('VidStats_viewCount')
VidStats_likeCountInd = vidDF.columns.get_loc('VidStats_likeCount')
VidStats_dislikeCountInd = vidDF.columns.get_loc('VidStats_dislikeCount')
VidStats_favoriteCountInd = vidDF.columns.get_loc('VidStats_favoriteCount')
VidStats_commentCountInd = vidDF.columns.get_loc('VidStats_commentCount')
```

```python
for i in range(0, len(vidDF.Output1)):
    
    vidDF.iat[i, UserInd] = ast.literal_eval(vidDF.iat[i, OutpInd]).get('User')
    vidDF.iat[i, VidDet_publishedAtInd] = parse(ast.literal_eval(vidDF.iat[i, OutpInd]).get('VideoDetails').get('publishedAt')).strftime('%Y-%m-%d %H:%M:%S %z')
    vidDF.iat[i, VidDet_titleInd] = ast.literal_eval(vidDF.iat[i, OutpInd]).get('VideoDetails').get('title')
    vidDF.iat[i, VidStats_viewCountInd] = ast.literal_eval(vidDF.iat[i, OutpInd]).get('VideoStats').get('viewCount')
    vidDF.iat[i, VidStats_likeCountInd] = ast.literal_eval(vidDF.iat[i, OutpInd]).get('VideoStats').get('likeCount')
    vidDF.iat[i, VidStats_dislikeCountInd] = ast.literal_eval(vidDF.iat[i, OutpInd]).get('VideoStats').get('dislikeCount')
    vidDF.iat[i, VidStats_favoriteCountInd] = ast.literal_eval(vidDF.iat[i, OutpInd]).get('VideoStats').get('favoriteCount')
    vidDF.iat[i, VidStats_commentCountInd] = ast.literal_eval(vidDF.iat[i, OutpInd]).get('VideoStats').get('commentCount')
```

```python
vidDF.rename(columns ={'Key1': 'KeyChannel', 'Key2': 'KeyVideo'}, inplace = True)
vidDF = vidDF.drop(['Output1'], axis = 1)
```

```python
vidDF = pd.concat([vidDF, vidDF.duplicated(keep = 'last', subset = ['KeyChannel', 'KeyVideo'])], axis = 1)
```

```python
vidDF.rename(columns ={0: 'DuplicateFlag'}, inplace = True)
```
