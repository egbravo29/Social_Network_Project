```python
OutpInd = twt.columns.get_loc('Output1')

CreatedAtInd = twt.columns.get_loc('CreatedAt')
in_reply_to_status_id_strInd = twt.columns.get_loc('in_reply_to_status_id_str')
retweet_countInd = twt.columns.get_loc('retweet_count')
favorite_countInd = twt.columns.get_loc('favorite_count')
commentTextInd = twt.columns.get_loc('CommentText')
commenterIdInd = twt.columns.get_loc('CommenterID')
commenterInd = twt.columns.get_loc('Commenter')
commenterLocationInd = twt.columns.get_loc('CommenterLocation')
commenterFollowers_countInd = twt.columns.get_loc('CommenterFollowers_count')
commenterFriends_countInd = twt.columns.get_loc('CommenterFriends_count')
```

```python
for i in range(0, len(twt.Output1)):
    
    twt.iat[i, CreatedAtInd] = parse(ast.literal_eval(twt.iat[i, OutpInd]).get('tweetContent').get('created_at')).strftime('%Y-%m-%d %H:%M:%S %z')
    twt.iat[i, in_reply_to_status_id_strInd] = ast.literal_eval(twt.iat[i, OutpInd]).get('tweetContent').get('in_reply_to_status_id_str')
    twt.iat[i, retweet_countInd] = ast.literal_eval(twt.iat[i, OutpInd]).get('tweetContent').get('retweet_count')
    twt.iat[i, favorite_countInd] = ast.literal_eval(twt.iat[i, OutpInd]).get('tweetContent').get('favorite_count')
    twt.iat[i, commentTextInd] = ast.literal_eval(twt.iat[i, OutpInd]).get('tweetContent').get('full_text')
    twt.iat[i, commenterIdInd] = ast.literal_eval(twt.iat[i, OutpInd]).get('tweetContent').get('userDict').get('id_str')
    twt.iat[i, commenterInd] = ast.literal_eval(twt.iat[i, OutpInd]).get('tweetContent').get('userDict').get('name')
    twt.iat[i, commenterLocationInd] = ast.literal_eval(twt.iat[i, OutpInd]).get('tweetContent').get('userDict').get('location')
    twt.iat[i, commenterFollowers_countInd] = ast.literal_eval(twt.iat[i, OutpInd]).get('tweetContent').get('userDict').get('followers_count')
    twt.iat[i, commenterFriends_countInd] = ast.literal_eval(twt.iat[i, OutpInd]).get('tweetContent').get('userDict').get('friends_count')
```

```python
twt = twt.drop(['Output1'], axis = 1)
```

```python
twt = twt.sort_values(by = ['Key1', 'Key2', 'CreatedAt', 'CommentID', 'in_reply_to_status_id_str', 'ExtractionDateTime'],
                            ascending = True).reset_index(drop = True)
twt = pd.concat([twt, twt.duplicated(keep = 'last', subset = ['Key1', 'Key2', 'CreatedAt', 'CommentID',
                                                              'in_reply_to_status_id_str'])], axis = 1)
twt.rename(columns ={0: 'DuplicateFlag'}, inplace = True)
```
